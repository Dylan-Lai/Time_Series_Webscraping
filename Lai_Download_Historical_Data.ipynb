```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
```

```python
# Create Ticker variables
data = yf.Ticker("DIS")
#Set the time range
hist_data = data.history(start=datetime.datetime(2010, 1, 1),end=datetime.datetime.today())
hist_data.insert(0,"Date",hist_data.index)
hist_data.head(20)
```

```python
#Visualization of the Closing price
plt.figure(figsize=(16,8))
plt.title("Closing Price History")
plt.plot(hist_data["Close"])
plt.xlabel("Date", fontsize=18)
plt.ylabel("Closing Price USD $", fontsize=18)
plt.show()
```

```python
hist_data.to_csv('Lai_Historical_Data.csv', index=True)
```
